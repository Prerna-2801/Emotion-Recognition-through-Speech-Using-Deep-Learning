```python
BASE_PATH      = '/content/drive/MyDrive/Datasets'
TESS_PATH      = f'{BASE_PATH}/tess/TESS Toronto emotional speech set data'
CREMA_PATH     = f'{BASE_PATH}/crema'

PROCESSED_PATH = f'{BASE_PATH}/processed_audio_v3'
MODEL_PATH     = f'{BASE_PATH}/ser_model_final.keras'
ENCODER_PATH   = f'{BASE_PATH}/label_encoder.pkl'

os.makedirs(PROCESSED_PATH, exist_ok=True)

print('TESS exists    :', os.path.exists(TESS_PATH))
print('CREMA-D exists :', os.path.exists(CREMA_PATH))

for name, path in [('TESS', TESS_PATH), ('CREMA', CREMA_PATH)]:
    n = len(glob.glob(path + '/**/*.wav', recursive=True))
    print(f'  {name}: {n} .wav files found')
```

```python
def label_tess(filename):
    """YAF_happy_01.wav → 'happy'"""
    name = filename.lower()
    for e in ['angry','happy','sad','fear','disgust','neutral','surprise','ps']:
        if e in name:
            return 'surprise' if e == 'ps' else e
    return 'neutral'

def label_crema(filename):
    """1001_DFA_ANG_XX.wav → 'angry'"""
    u = filename.upper()
    crema_map = {'_ANG_':'angry','_HAP_':'happy','_SAD_':'sad',
                 '_FEA_':'fear', '_DIS_':'disgust','_NEU_':'neutral'}
    for k, v in crema_map.items():
        if k in u: return v
    return 'neutral'

assert label_tess('YAF_happy_01.wav')            == 'happy'
assert label_tess('OAF_Fear_01.wav')             == 'fear'
assert label_crema('1001_DFA_SAD_XX.wav')        == 'sad'
assert label_crema('1001_IEO_NEU_XX.wav')        == 'neutral'
print('All label parser tests passed ')
```

```python
SR      = 16000
N_MELS  = 128
N_FFT   = 1024
HOP_LEN = 256
MAX_LEN = 128


def denoise_audio(y, sr):
    """
    FIX #6: Remove background noise using spectral gating.
    Uses first 0.5s as noise profile estimate.
    """
    noise_sample = y[:int(sr * 0.5)]
    if len(noise_sample) > 0:
        y_denoised = nr.reduce_noise(y=y, sr=sr, y_noise=noise_sample,
                                     prop_decrease=0.75, stationary=False)
        return y_denoised
    return y


def pad_or_crop(S, max_len=MAX_LEN):
    """
    FIX #1: Proper fixed-length. Zero-pad on right if short.
    Crop from centre if too long (keeps the emotionally dense middle).
    Original bug: np.resize() wraps pixel values → fake patterns!
    """
    if S.shape[1] < max_len:
        pad_width = max_len - S.shape[1]
        S = np.pad(S, ((0, 0), (0, pad_width)), mode='constant', constant_values=0)
    else:
        start = (S.shape[1] - max_len) // 2
        S = S[:, start : start + max_len]
    return S


def normalize_spectrogram(S):
    """
    FIX #7: Use mean/std normalization (z-score) instead of max-divide.
    More stable across varying loudness levels between datasets.
    """
    mean = S.mean()
    std  = S.std() + 1e-8
    return (S - mean) / std


def extract_mel(file_path):
    """
    Full fixed pipeline for one audio file.
    Returns shape: (128, 128) — float32
    """

    y, _ = librosa.load(file_path, sr=SR, mono=True)

    y = denoise_audio(y, SR)

    y, _ = librosa.effects.trim(y, top_db=30)

    if len(y) < 512:
        y = np.pad(y, (0, 512 - len(y)))

    S = librosa.feature.melspectrogram(
        y=y, sr=SR, n_mels=N_MELS, n_fft=N_FFT, hop_length=HOP_LEN
    )

    log_S = librosa.power_to_db(S, ref=np.max)

    log_S = pad_or_crop(log_S, MAX_LEN)

    log_S = normalize_spectrogram(log_S)

    return log_S.astype(np.float32)

test_wavs = glob.glob(TESS_PATH + '/**/*.wav', recursive=True)
if test_wavs:
    test_mel = extract_mel(test_wavs[0])
    print(f'Feature shape  : {test_mel.shape}')
    print(f'Value range    : [{test_mel.min():.3f}, {test_mel.max():.3f}]')
    print(f'Mean / Std     : {test_mel.mean():.3f} / {test_mel.std():.3f}')
    print('extract_mel() OK ')
else:
    print('  No wavs found — check TESS_PATH')
```

```python
def process_and_save(dataset_path, label_fn, x_out, y_out, force=False):
    """Extract fixed features for all .wav files. Saves .npy to Drive."""
    if os.path.exists(x_out) and not force:
        print(f'  Already done → {os.path.basename(x_out)}')
        return (np.load(x_out),
                np.load(y_out, allow_pickle=True))

    wavs = glob.glob(dataset_path + '/**/*.wav', recursive=True)
    print(f'  Found {len(wavs)} wav files')

    X_list, y_list = [], []
    errors = 0
    for fp in tqdm(wavs, desc=os.path.basename(dataset_path)):
        label = label_fn(os.path.basename(fp))
        try:
            mel = extract_mel(fp)
            X_list.append(mel)
            y_list.append(label)
        except Exception as e:
            errors += 1

    print(f'  Skipped {errors} corrupt/empty files')
    X = np.array(X_list, dtype=np.float32)
    y = np.array(y_list)
    np.save(x_out, X)
    np.save(y_out, y)
    print(f'  Saved → shape {X.shape}  labels: {Counter(y)}')
    return X, y


print('━'*50)
print('\nProcessing TESS...')
X_tess, y_tess = process_and_save(
    TESS_PATH, label_tess,
    f'{PROCESSED_PATH}/X_tess.npy',
    f'{PROCESSED_PATH}/y_tess.npy'
)

print('\nProcessing CREMA-D...')
X_crema, y_crema = process_and_save(
    CREMA_PATH, label_crema,
    f'{PROCESSED_PATH}/X_crema.npy',
    f'{PROCESSED_PATH}/y_crema.npy'
)

print('\n All datasets processed!')
print(f'TESS    : {X_tess.shape}')
print(f'CREMA-D : {X_crema.shape}')
```

```python
import numpy as np
import pickle
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from collections import Counter

PROCESSED_PATH = f'{BASE_PATH}/processed_audio_v3'

X_tess    = np.load(f'{PROCESSED_PATH}/X_tess.npy')
y_tess    = np.load(f'{PROCESSED_PATH}/y_tess.npy',    allow_pickle=True)
X_crema   = np.load(f'{PROCESSED_PATH}/X_crema.npy')
y_crema   = np.load(f'{PROCESSED_PATH}/y_crema.npy',   allow_pickle=True)

print('Loaded:')
print(f'  TESS    : {X_tess.shape}  {Counter(y_tess)}')
print(f'  CREMA-D : {X_crema.shape}  {Counter(y_crema)}')

X_total = np.concatenate([X_tess, X_crema], axis=0)
y_total = np.concatenate([y_tess, y_crema], axis=0)
print(f'\nCombined : {X_total.shape}  {X_total.dtype}')
print(f'Labels   : {Counter(y_total)}')

le = LabelEncoder()
y_total_encoded = le.fit_transform(y_total)
print(f'\nClasses  : {list(enumerate(le.classes_))}')
print(f'y dtype  : {y_total_encoded.dtype}  (must be int64, not string!)')

with open(ENCODER_PATH, 'wb') as f:
    pickle.dump(le, f)
print('Label encoder saved ✓')

X_train, X_test, y_train, y_test = train_test_split(
    X_total, y_total_encoded,
    test_size=0.20, random_state=42, stratify=y_total_encoded
)
print(f'\nTrain : {X_train.shape}  {Counter(y_train)}')
print(f'Test  : {X_test.shape}   {Counter(y_test)}')
```

```python
X_train = X_train[..., np.newaxis]
X_test  = X_test[...,  np.newaxis]
print('X_train :', X_train.shape)
print('X_test  :', X_test.shape)
```

```python
print('Before balancing:', Counter(y_train))

counter   = Counter(y_train)
max_count = max(counter.values())

balanced_X, balanced_y = [], []

for label in sorted(counter.keys()):
    idx    = np.where(y_train == label)[0]
    cls_X  = X_train[idx]
    cls_y  = y_train[idx]

    repeats   = max_count // len(cls_X)
    remainder = max_count %  len(cls_X)

    bal_X = np.concatenate([cls_X] * repeats, axis=0)
    bal_y = np.concatenate([cls_y] * repeats, axis=0)

    if remainder > 0:
        extra_idx = np.random.choice(len(cls_X), remainder, replace=True)
        bal_X = np.concatenate([bal_X, cls_X[extra_idx]], axis=0)
        bal_y = np.concatenate([bal_y, cls_y[extra_idx]], axis=0)

    balanced_X.append(bal_X)
    balanced_y.append(bal_y)

X_train_bal = np.concatenate(balanced_X, axis=0)
y_train_bal = np.concatenate(balanced_y, axis=0)

perm        = np.random.permutation(len(X_train_bal))
X_train_bal = X_train_bal[perm]
y_train_bal = y_train_bal[perm]

print('After balancing :', Counter(y_train_bal))
print('Balanced X shape:', X_train_bal.shape)
print('Balanced y dtype:', y_train_bal.dtype, '← must be int64, not string')
print('Balanced y range:', y_train_bal.min(), '–', y_train_bal.max())
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

@tf.keras.utils.register_keras_serializable(package='SER')
class PositionalEncoding(layers.Layer):
    def __init__(self, num_positions=256, d_model=128, **kwargs):
        super().__init__(**kwargs)
        self.num_positions = num_positions
        self.d_model       = d_model

        pos  = tf.cast(tf.range(num_positions)[:, tf.newaxis], tf.float32)
        dim  = tf.cast(tf.range(d_model)[tf.newaxis, :],       tf.float32)
        rate = 1.0 / tf.pow(10000.0, (2.0 * tf.floor(dim/2.0)) / tf.cast(d_model, tf.float32))
        ang  = pos * rate

        sin_e = tf.sin(ang[:, 0::2])
        cos_e = tf.cos(ang[:, 1::2])
        pe    = tf.reshape(tf.stack([sin_e, cos_e], axis=-1), (num_positions, d_model))
        self.pe = tf.cast(pe[tf.newaxis, ...], tf.float32)

    def call(self, x):
        return x + self.pe[:, :tf.shape(x)[1], :]

    def get_config(self):
        cfg = super().get_config()
        cfg.update({'num_positions': self.num_positions, 'd_model': self.d_model})
        return cfg


def build_cnn_transformer(input_shape=(128, 128, 1), num_classes=8,
                           num_heads=4, transformer_layers=3,
                           ffn_units=256, dropout=0.3):
    inputs = layers.Input(shape=input_shape)

    x = layers.Conv2D(32,  3, padding='same')(inputs)
    x = layers.BatchNormalization()(x); x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D(2)(x)

    x = layers.Conv2D(64,  3, padding='same')(x)
    x = layers.BatchNormalization()(x); x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D(2)(x)

    x = layers.Conv2D(128, 3, padding='same')(x)
    x = layers.BatchNormalization()(x); x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D(2)(x)

    x = layers.Conv2D(128, 3, padding='same')(x)
    x = layers.BatchNormalization()(x); x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D(2)(x)

    seq_len = x.shape[1] * x.shape[2]
    d_model = x.shape[3]
    x = layers.Reshape((seq_len, d_model))(x)
    x = PositionalEncoding(num_positions=seq_len, d_model=d_model)(x)
    x = layers.Dropout(dropout)(x)

    for _ in range(transformer_layers):
        attn = layers.MultiHeadAttention(
            num_heads=num_heads, key_dim=d_model // num_heads, dropout=dropout)(x, x)
        x    = layers.LayerNormalization(epsilon=1e-6)(x + attn)

        ff   = layers.Dense(ffn_units, activation='gelu')(x)
        ff   = layers.Dropout(dropout)(ff)
        ff   = layers.Dense(d_model)(ff)
        x    = layers.LayerNormalization(epsilon=1e-6)(x + ff)

    x       = layers.GlobalAveragePooling1D()(x)
    x       = layers.Dense(256, activation='gelu')(x)
    x       = layers.Dropout(dropout)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    return models.Model(inputs, outputs, name='SER_CNN_Transformer')


NUM_CLASSES = len(le.classes_)
model = build_cnn_transformer(num_classes=NUM_CLASSES)
model.summary(line_length=80)
```

```python
import pickle
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

callbacks = [
    EarlyStopping(
        monitor='val_accuracy', patience=8,
        restore_best_weights=True, verbose=1
    ),
    ModelCheckpoint(
        MODEL_PATH, monitor='val_accuracy',
        save_best_only=True, verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss', factor=0.5,
        patience=4, min_lr=1e-6, verbose=1
    )
]

history = model.fit(
    X_train_bal, y_train_bal,
    validation_split=0.10,
    epochs=60,
    batch_size=32,
    callbacks=callbacks
)

with open(f'{BASE_PATH}/training_history.pkl', 'wb') as f:
    pickle.dump(history.history, f)
print('\nTraining complete — history saved ')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

test_loss, test_acc = model.evaluate(X_test, y_test, verbose=1)
print(f'\n Test Accuracy: {test_acc*100:.2f}%')

y_pred_prob   = model.predict(X_test)
y_pred_labels = np.argmax(y_pred_prob, axis=1)

print('\n' + classification_report(
    y_test, y_pred_labels,
    target_names=le.classes_,
    zero_division=0
))
```

```python
cm = confusion_matrix(y_test, y_pred_labels)
plt.figure(figsize=(9, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Confusion Matrix — SER (Fixed Preprocessing)', fontsize=13)
plt.xlabel('Predicted'); plt.ylabel('Actual')
plt.tight_layout()
plt.savefig(f'{BASE_PATH}/confusion_matrix_v3.png', dpi=150)
plt.show()
```
